# Pandas Series - Like a Table Column - 1D Array holding any type of Data

```python
my_series=[5,9,12,27]
p1 = pd.Series(my_series)
print(p1)
```

```python
p1[2]
```

```python
my_index=["A","B","C","D"]
p2=pd.Series(my_series,my_index)
print(p2)
```

```python
p2["B"]
```

```python
cars={"Tesla":12,"Mercedes":42,"BMW":3}
p3=pd.Series(cars)
print(p3)
```

```python
p3["Mercedes"]
```
